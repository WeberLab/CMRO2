```r
data <- read.csv("JillDeVis_Impact_of_neonate_haematocrit.csv")
```

```r
head(data)
```

```r
colnames(data) <- c("PNA", "T1b")
```

```r
head(data)
```

```r
plot(T1b ~ PNA, data = data)
```

```r
install.packages("ggpubr")
```

```r
library("ggpubr")
ggscatter(data, x = "PNA", y = "T1b", 
          add = "reg.line", conf.int = TRUE, 
          cor.coef = TRUE, cor.method = "pearson",
          xlab = "PNA (days)", ylab = "T1b")
```

```r
model <- lm(T1b ~ PNA, data = data)
```
